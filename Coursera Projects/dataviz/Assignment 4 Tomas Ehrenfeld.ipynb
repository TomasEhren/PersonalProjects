Data Sources:  

* Unemployment Rate in Ann Arbor, MI (MSA):  https://fred.stlouisfed.org/series/ANNA426URN 
* Total Real Gross Domestic Product for Ann Arbor, MI (MSA):  https://fred.stlouisfed.org/series/RGMP11460
* Per Capita Personal Income in Ann Arbor, MI (MSA): https://fred.stlouisfed.org/series/ANNA426PCPI
* Average Weekly Wages for Employees in State Government Establishments in Ann Arbor, MI (MSA): https://fred.stlouisfed.org/series/ENUC114640210

Let's first import the libreries and load de datasets

```python
import pandas as pd
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
plt.style.use('seaborn-colorblind')
import seaborn as sns
import datetime as dt
from IPython.display import display_html # To make side by side outputs
```

This little code we'll allow us to make side by side output for better redability of our DataFrames on the next cell.

```python
def display_side_by_side(*args):
    html_str=''
    for df in args:
        html_str+=df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'),raw=True)
```

```python
GDP = pd.read_csv('GDP.csv')
non_gob_wages = pd.read_csv('per_capita_income.csv')
gob_wages = pd.read_csv('gobernment_wages.csv')
unemploy = pd.read_csv('uneemployment.csv')
```

Because we are looking at time serieses, we'll have to make an actual date out of the "DATE" column, and it'll also be useful to make this our index. Let's also sort by the newly set index, just in case data isn't ordered.

```python
GDP = GDP.set_index(pd.to_datetime(GDP.DATE))\
        .sort_index()\
        .drop(columns=['DATE'])\
        .rename(columns={'RGMP11460': 'GDP'})

non_gob_wages = non_gob_wages.set_index(pd.to_datetime(non_gob_wages.DATE))\
                            .sort_index()\
                            .drop(columns=['DATE'])\
                            .rename(columns={'ANNA426PCPI': 'non_gob_wages'})

gob_wages = gob_wages.set_index(pd.to_datetime(gob_wages.DATE))\
                    .sort_index()\
                    .drop(columns=['DATE'])\
                    .rename(columns={'ENUC114640210': 'gob_wages'})


unemploy = unemploy.set_index(pd.to_datetime(unemploy.DATE))\
                    .sort_index()\
                    .drop(columns=['DATE'])\
                    .rename({'ANNA426URN': 'unemployment'})

unemploy.columns = ['unemployment'] #borrar
```

Let's look into each DataFrame to see what it looks like.

```python
display_side_by_side(GDP.head(10), non_gob_wages.head(10), gob_wages.head(10), unemploy.head(10))
display_side_by_side(GDP.tail(10), non_gob_wages.tail(10), gob_wages.tail(10), unemploy.tail(10))
```
